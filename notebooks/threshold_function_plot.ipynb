# Threshold Function Plot

```python
import numpy as np
from scipy import special
import math
import torch
import torch.nn.functional as F

import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
threshold = 7.5
slope = 2.0
x = torch.linspace(0, 20, 101)
y = 1.0 - torch.sigmoid(slope * (x-threshold))

plt.plot(x, y);
plt.grid(True);
```
