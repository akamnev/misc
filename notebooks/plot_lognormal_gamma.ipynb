# Построение графика LogNormal - Gamma

```python
import numpy as np
from scipy import special
import math


import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def const_coeff(alpha, beta):
    return special.loggamma(alpha) - alpha * math.log(beta) - 0.5 * math.log(2*math.pi) - 0.5

def kld(alpha, beta, sigma, mu):
    c0 = const_coeff(alpha, beta)
    c1 = -np.log(sigma) + beta * np.exp(mu + 0.5 * sigma**2) - alpha * mu
    return c0 + c1
```

```python
alpha = 0.5
beta = 0.01

mu = 0.0
sigma = np.linspace(0.001, 3, 1000)

y = kld(alpha, beta, sigma, mu)

plt.plot(sigma, y)
```

```python
alpha = 0.2
beta = 0.1

mu = np.linspace(0.001, 3, 1000)
sigma = 1.0

y = kld(alpha, beta, sigma, mu)

plt.plot(mu, y)
```

```python
sigma = 1.0
mu = -0.5 * sigma**2 + 0.0125 * 0
a = 10.0

x = np.random.normal(size=(10000))
t = 3.0
m = x > t 
x[m] = x[m] % 2

x = mu + sigma*x 
x = np.exp(x)
print(np.mean(x))
plt.hist(x, bins=100);
```

```python
rho = 4.0
lam = 0.0

mu = 1.0
sigma = np.linspace(0.001, 20, 1000)

kld = 0.5*(sigma**2/rho**2 + (mu - lam)**2/rho**2 - 1.0 - np.log(sigma**2/rho**2))

plt.plot(sigma, kld)
plt.grid(True);
plt.xlabel('$\sigma$')
plt.title(r'Gaussian KLD for $\rho=1.0$, $\lambda=0.0$, $\mu=1.0$');
```
