# Построение графика Gaussian KLD

```python
import numpy as np
from scipy import special
import math


import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
rho = 4.0
lam = 0.0

mu = 1.0
sigma = np.linspace(0.001, 20, 1000)

kld = 0.5*(sigma**2/rho**2 + (mu - lam)**2/rho**2 - 1.0 - np.log(sigma**2/rho**2))

plt.plot(sigma, kld)
plt.grid(True);
plt.xlabel('$\sigma$')
plt.title(r'Gaussian KLD for $\rho=1.0$, $\lambda=0.0$, $\mu=1.0$');
```
