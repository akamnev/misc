# Activation function plot

```python
import numpy as np
from scipy import special
import math
import torch
import torch.nn.functional as F

import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
np.log(8.0), np.log(0.125)
```

```python
p = torch.exp(torch.linspace(math.log(0.125), math.log(8.0), 10))
p = 2*math.pi* p
plt.plot(p)
```

```python
x = np.linspace(0, 1, 1000)
for pi in p:
    y = np.sin(x*float(pi))
    plt.plot(x, y);
plt.grid(True);
```

```python
alpha = 0.1
b = 1.0
x = np.linspace(-10, 10, 1000)
x *= b
f = np.log(1.0 + np.exp(x)) - np.log(1.0 + np.exp(-alpha * x))
f /= b
x /= b

plt.plot(x, f)
plt.grid(True);
plt.title(r'log sum for $\alpha=0.1$, $\beta=1.0$');
```

```python
alpha = 0.01
beta = 10.0
x = torch.linspace(-100, 1, 1000)
bx = beta * x
y = -F.logsigmoid(-bx) + F.logsigmoid(alpha * bx)
y /= beta

plt.plot(x, y);
plt.grid(True);
plt.title(fr'log sum for $\alpha={alpha}$, $\beta={beta}$');
```

```python
a, b = 10.0, 0.01
x = np.linspace(-5, 1, 1000)
f = (1-b)/(1.0 + np.exp(-a*x)) + b
f = f * x

plt.plot(x, f)
plt.grid(True);
plt.title(r'$x \cdot sigma(x)$ for $\alpha=0.1$, $\beta=10.0$');
```
